# Drugs Classification (Logistic Regression)

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('drug200.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## One Hot Encoding and Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
X[:, 1] = le.fit_transform(X[:, 1])
X[:, 2] = le.fit_transform(X[:, 2])
X[:, 3] = le.fit_transform(X[:, 3])
```

```python
print(X)
```

```python
print(y)
```

## Spliting dataset to train and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
print(len(X_train))
```

```python
print(len(X_test))
```

```python
print(len(y_train))
```

```python
print(len(y_test))
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, [0, 4]] = sc.fit_transform(X_train[:, [0, 4]])
X_test[:, [0, 4]] = sc.fit_transform(X_test[:, [0, 4]])
```

```python
print(X_train)
```

```python
print(X_test)
```

## Data Fitting and Model Creation

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Prediction

```python
y_pred = classifier.predict(X_test)
```

```python
print(y_pred)
```

## Metrics

```python
from sklearn.metrics import multilabel_confusion_matrix
mlc = multilabel_confusion_matrix(y_test, y_pred)
print(mlc)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred) * 100
```
